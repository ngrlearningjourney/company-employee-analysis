## **Which Column Has the Highest Correlation in Predicting Whether an Employee Will Leave?**

```python
from sklearn.preprocessing import LabelEncoder
# Initialize LabelEncoder
label_encoder = LabelEncoder()

employee_df['LeaveOrNot'] = label_encoder.fit_transform(employee_df['LeaveOrNot'])
numerical_columns = employee_df[['Age','ExperienceInCurrentDomain','JoiningYear']]


from scipy.stats import pointbiserialr

correlations = []
for column in numerical_columns:
  correlation, p_value = pointbiserialr(employee_df[column], employee_df['LeaveOrNot'])
  correlations.append(correlation)

data = {
    'numerical_columns': numerical_columns.columns,
    'correlations': correlations
}
numerical_columns_correlations = pd.DataFrame(data)
numerical_columns_correlations.sort_values(by='correlations', ascending=False)
```

**Correlation Summary with `LeaveOrNot`**

**Key Points:**
1. **Age** shows a weak positive correlation (*0.114943*) with `LeaveOrNot`, suggesting that as age increases, there is a slight tendency for employees to leave.
2. **ExperienceInCurrentDomain** has a very weak positive correlation (*0.021181*) with `LeaveOrNot`, indicating little to no relationship between experience in the current domain and leaving.
3. **JoiningYear** has a weak negative correlation (*-0.150650*), meaning employees who joined earlier are slightly less likely to leave.

While these correlations are weak, they may still provide insights into employee behavior trends.

```python
nominal_columns = ['City','Gender', 'EverBenched']

# Fit and transform the categorical column to numeric values
for column in ordinal_columns:
  employee_df[column] = label_encoder.fit_transform(employee_df[column])

employee_df['Education'] = label_encoder.fit_transform(employee_df['Education'])
```

```python
from scipy.stats import chi2_contingency
import pandas as pd

chiResult = []
pValue = []
cramer_v = []

for column in nominal_columns:
    # Create a contingency table
    contingency_table = pd.crosstab(employee_df[column], employee_df['LeaveOrNot'])

    # Perform Chi-Square Test (unpacking the 4 values)
    chi2, p, dof, expected = chi2_contingency(contingency_table)

    # Calculate Cram√©r's V
    n = contingency_table.to_numpy().sum()  # Total number of observations
    min_dim = min(contingency_table.shape) - 1  # Minimum dimension minus 1
    cramers_v = np.sqrt(chi2 / (n * min_dim))

    # Append the chi2 and p values to their respective lists
    chiResult.append(chi2)
    pValue.append(p)
    cramer_v.append(cramers_v)

# Prepare the results in a DataFrame
data = {
    'nominal_columns': nominal_columns,
    'Chi-Square Statistic': chiResult,
    'P-value': pValue,
    'Cramer v' : cramer_v
}
numerical_columns_correlations = pd.DataFrame(data)
numerical_columns_correlations
```

**Chi-Square Test and Cram√©r's V Summary**

| Column         | Chi-Square Statistic | P-value          | Cram√©r's V | Interpretation                              |
|----------------|----------------------|------------------|------------|----------------------------------------------|
| **City**       | 63.46               | 1.66e-14         | 0.1515     | Significant weak to moderate correlation     |
| **Gender**     | 99.48               | 1.98e-23         | 0.1897     | Significant weak to moderate correlation     |
| **EverBenched**| 3.59                | 0.058            | 0.0360     | Not significant; negligible correlation      |

**Key Points:**
1. **City** and **Gender** show statistically significant relationships with `LeaveOrNot` (p-value < 0.05).
2. Both `City` and `Gender` exhibit weak to moderate associations (Cram√©r's V < 0.3).
3. **EverBenched** does not show a significant relationship with `LeaveOrNot` (p-value > 0.05) and has a negligible association (Cram√©r's V = 0.036).


```python
from scipy.stats import kendalltau
ordinal_columns = ['Education', 'PaymentTier']
tau_values = []
p_values = []
for column in ordinal_columns:
  # Calculate Kendall's Tau-b correlation
  tau, p_value = kendalltau(employee_df[column], employee_df['LeaveOrNot'])
  tau_values.append(tau)
  p_values.append(p_value)

# Prepare the results in a DataFrame
data = {
    'ordinal_columns': ordinal_columns,
    'Kendall Tau': tau_values,
    'P-value': p_values,
}

ordinal_columns_correlations = pd.DataFrame(data)
ordinal_columns_correlations
```

**Kendall's Tau Correlation Summary**

| Column         | Kendall Tau | P-value          | Interpretation                              |
|----------------|-------------|------------------|---------------------------------------------|
| **Education**  | -0.0454     | 1.46e-02         | Weak negative correlation, statistically significant |
| **PaymentTier**| 0.1557      | 4.09e-17         | Weak positive correlation, statistically significant |

**Key Points:**
1. Both **Education** and **PaymentTier** have statistically significant correlations with `LeaveOrNot` (p-value < 0.05).
2. **Education** shows a weak negative correlation, meaning as education level changes, there is a slight tendency for the likelihood of `LeaveOrNot` to move inversely.
3. **PaymentTier** shows a weak positive correlation, indicating that higher payment tiers are slightly associated with increased likelihood of `LeaveOrNot`.

# Classification

```python
# Replace specific categorical values in the DataFrame with numerical codes
# This is useful for preparing the data for machine learning models that require numerical inputs
employee_df = employee_df.replace(
    [
        'Bachelors', 'Masters', 'PHD',  # Education levels
        'Bangalore', 'Pune', 'New Delhi',  # Locations
        'Male', 'Female',  # Gender
        'No', 'Yes',  # Binary responses
        'Not Leave', 'Leave'  # Employee leave status
    ],
    [
        1, 2, 3,  # Numerical codes for education levels
        1, 2, 3,  # Numerical codes for locations
        1, 2,  # Numerical codes for gender
        1, 2,  # Numerical codes for binary responses
        1, 2   # Numerical codes for leave status
    ]
)
```

```python
X = employee_df.iloc[:,:-1].values
y = employee_df.iloc[:,-1].values
# x is independent variable and y is dependent variable
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.1)
# Split the dataset into training and testing sets
# X: Features (input data)
# y: Target (output labels)
# test_size: Proportion of the dataset to include in the test split (10% in this case)
# X_train, X_test: Training and testing features
# y_train, y_test: Training and testing labels
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
# StandardScaler standardizes the data by removing the mean and scaling to unit variance
sc = StandardScaler()

# Fit the scaler to the training data and transform it
# This calculates the mean and standard deviation from X_train and applies scaling
X_train = sc.fit_transform(X_train)

# Transform the test data using the same scaling parameters calculated from the training data
# to ensure consistency between training and test data scaling.
X_test = sc.transform(X_test)
```

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression

# Initialize the Logistic Regression classifier
# random_state: Ensures reproducibility by setting a fixed random seed
classifier = LogisticRegression(random_state=0)

# Fit the Logistic Regression model to the training data
# X_train: Training features
# y_train: Training labels
classifier.fit(X_train, y_train)
```

```python
# Predict the labels for the test set using the trained classifier
# X_test: Features of the test set
# y_pred: The predicted labels for the test data
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

# Generate the confusion matrix to evaluate the performance of the classification model
# y_test: Actual labels of the test set
# y_pred: Predicted labels from the model
cm = confusion_matrix(y_test, y_pred)

# Print the confusion matrix to show the model's performance
print("Confusion Matrix\n", cm)

# Calculate and print the accuracy score, which is the proportion of correct predictions
# accuracy_score compares the actual and predicted labels
print("Accuracy score", accuracy_score(y_test, y_pred))
```

The accuracy score of 0.7292418772563177 (approximately 72%) is the proportion of correct predictions made by the model.

**KNN**

```python
from sklearn.neighbors import KNeighborsClassifier

# Initialize the K-Nearest Neighbors (KNN) classifier
# n_neighbors: Number of neighbors to consider when making a prediction (5 in this case)
# metric: The distance metric to use for calculating the distance between points ("euclidean" calculates the straight-line distance)
# p: Power parameter for the Minkowski distance metric (p=2 corresponds to Euclidean distance)
classifier = KNeighborsClassifier(n_neighbors=5, metric="euclidean", p=2)

# Fit the KNN model to the training data
# X_train: Training features
# y_train: Training labels
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

# Generate the confusion matrix to evaluate the performance of the KNN model
# y_test: Actual labels of the test set
# y_pred: Predicted labels from the KNN model
cm2 = confusion_matrix(y_test, y_pred)

# Print the confusion matrix to see the breakdown of the model's performance
print("Confusion Matrix\n", cm2)

# Calculate and print the accuracy score for the KNN model
# accuracy_score compares the actual labels (y_test) with the predicted labels (y_pred)
print("Accuracy Score: ", accuracy_score(y_test, y_pred))
```

Accuracy: 0.7509025270758123 (approximately 75%) The accuracy score indicates that the model correctly predicted the class labels 74% of the time.

**Recommendations**

1. **Improve Retention Strategies for Master's Degree Holders**  
   - Conduct exit interviews and satisfaction surveys to identify reasons for attrition among Master's degree holders.  
   - Offer tailored growth opportunities, such as leadership training or advanced technical skills, to better engage and retain them.  

2. **Address Domain Experience Gaps**  
   - Provide structured onboarding and training programs to upskill employees with no prior domain experience, particularly among Bachelor's degree holders.  
   - Develop mentorship programs where experienced employees guide newcomers.  

3. **Targeted Retention for Experienced Employees**  
   - Introduce incentives like promotions, salary increases, and career development plans for employees with more than 3 years of experience.  
   - Foster a strong organizational culture to encourage long-term commitment.  

4. **Optimize Bench Management**  
   - Reevaluate hiring practices to prevent over-hiring, especially for roles requiring Bachelor's degrees.  
   - Assign benched employees to short-term projects, cross-training opportunities, or skill development programs to enhance their utilization.  

5. **Payment Tier Utilization Strategy**  
   - Investigate the reasons for higher benching in Payment Tier 3 and focus on aligning hiring with actual workforce needs.  
   - Create pathways for employees in lower tiers to advance through internal training and certification programs, reducing benching rates.
